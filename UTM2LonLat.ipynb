```python
import utm
import pandas as pd
```

```python
filepath="C:\\Users\\kuion\\OneDrive\\Dokumendid\\MadridMapConv\\MADRID ZBE. Photo-red devices located on roads inside the M-30.tsv"
```

```python
def utm2LonLat(row):
    # print(row.index)
    # print(utm.to_latlon(row["UTM_X"], row["UTM_Y"], 30, 'T'))
    lat, lon = utm.to_latlon(row["UTM_X"], row["UTM_Y"], 30, 'T')
    row["LATITUD"] = lat
    row["LONGITUD"] = lon
    return row
```

```python
df = pd.read_csv(filepath_or_buffer=filepath, delimiter="\t", thousands=',')
# print(df)
df = df.apply(utm2LonLat, axis=1)
# print(df)
df#["LATITUD"]
```

```python
df.to_csv(filepath[:-3]+"csv", index=False)
```
